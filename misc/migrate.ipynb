```python
import pandas as pd

# Specify the path to the Excel file
excel_file_path = "data.xlsx"

# Read the Excel file into a pandas DataFrame with row 0 as the header
df = pd.read_excel(excel_file_path, header=1)


df = df.fillna('')


df = df[df['Autor'] != ""]


print(df)
```

```python
import requests

# Set the base URL of your Strapi backend
base_url = "http://localhost:1337"

api_key = ""
```

```python

unique_authors = df['Autor'].unique()



# Iterate over the "Autor" column
for author_name in unique_authors:
    # Define the data to send
    headers = {
        "Content-Type": "application/json",
        "Authorization": f"Bearer {api_key}"
    }


    data ={
        "data": {
            "name": author_name,
        }
    }

    # Send a POST request to the Strapi API
    response = requests.post(base_url + "/api/artists", json=data, headers=headers)

    # Check the response
    if response.status_code == 200:
        print(f'Successfully added author {author_name}')
    else:
        print(f'Failed to add author {author_name}. Response: {response.text}')
```

```python
import requests



# Set the headers for the request
headers = {
    "Content-Type": "application/json",
    "Authorization": f"Bearer {api_key}"
}

# Send a GET request to the authors endpoint
response = requests.get(base_url + "/api/artists?pagination[page]=1&pagination[pageSize]=90", headers=headers)

# Check the response
if response.status_code == 200:
    artists = response.json()
    artists = {artist["attributes"]['name']: artist['id'] for artist in artists["data"]}
    print(artists)
else:
    print(f'Failed to get authors. Response: {response.text}')
```

```python
import requests



# Iterate over the rows of the dataframe
for index, row in df.iterrows():
    # Get the values from the row
    song_name = row['Titel']
    artist_id = artists[row['Autor']]
    tags = row['Thema']
    
    if row['Altersgruppe'] == 'Alle' or row['Altersgruppe'] == '?' or row['Altersgruppe'] == '':
        min_age = 0
        max_age = 100
    else:
        min_age, rest = row['Altersgruppe'].split('-')
        max_age = rest.split(' ')[0]
        min_age = int(min_age)
        max_age = int(max_age)


    headers = {
        "Content-Type": "application/json",
        "Authorization": f"Bearer {api_key}"
    }


    data ={
        "data": {
            "title": song_name,
            "artists": [artist_id],
            "tags": tags,
            "file": 1,
            "minAge": min_age,
            "maxAge": max_age,

        }
    }

    # Send a POST request to the Strapi API
    response = requests.post(base_url + "/api/songs", json=data, headers=headers)

    # Check the response
    if response.status_code == 200:
        print(f'Successfully added Song {song_name}')
    else:
        print(f'Failed to add Song {song_name}. Response: {response.text}')
```
